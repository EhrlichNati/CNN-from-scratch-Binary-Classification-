```python
train_set_x_orig, train_set_y_orig, test_set_x_orig, test_set_y_orig, classes = cat_data_loader()
x_train, y_train, x_test, y_test, classes = flatten_data(train_set_x_orig, train_set_y_orig, test_set_x_orig, test_set_y_orig, classes)

print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)


```

```python
layers = [x_train.shape[0], 8, 1]
#after_train_parameters = train_model(x_train,y_train, layers=layers ,learning_rate=0.005,train_iterations=1500,cost_function=max_entropy_cost)
def net_init(layers):

    layers_parameters = {}
    L = len(layers)

    for layer_num in range(1, L):
        layers_parameters['weights_matrix ' + str(layer_num)] = np.random.rand(layers[layer_num], layers[layer_num-1])*0.01
        layers_parameters['bias_v ' + str(layer_num)] = np.zeros((layers[layer_num], 1))

    return layers_parameters

parameters = net_init(layers)
```

```python
def forward_propagate(activation_prev, weights_matrix, bias_v, function):
    """Calculate linear output"""
    
    linear_output = np.dot(weights_matrix, activation_prev) + bias_v


    print(activation_prev, weights_matrix.tolist(), bias_v.tolist())
    print("linear_output",linear_output.tolist())
    print(linear_output.shape)


    """Choose activation"""
    if function == "sigmoid":
        activation_output, activation_cache = sigmoid(linear_output)
    elif function == "relu":
        activation_output, activation_cache = relu(linear_output)

    linear_cache = activation_prev, weights_matrix, bias_v


    return activation_output, linear_cache, activation_cache




def L_forward_propagate(X, parameters, layers):
    L = len(parameters)//2
    activation_prev = X
    caches = []


    for layer in range(1, L):
        weights_matrix = parameters['weights_matrix ' + str(layer)]
        bias_v = parameters['bias_v ' + str(layer)]
        
        
        activation_output, linear_cache, activation_cache = forward_propagate(activation_prev, weights_matrix, bias_v, "relu")
        caches.append((activation_output, linear_cache, activation_cache))
        activation_prev = activation_output

    """Propagate to L(final) layer, with different activation function"""
    weights_matrix = parameters['weights_matrix ' + str(L)]
    bias_v = parameters['bias_v ' + str(L)]
    activation_output, linear_cache, activation_cache = forward_propagate(activation_prev, weights_matrix, bias_v, "sigmoid")
    caches.append((activation_output, linear_cache, activation_cache))

    y_predict = activation_output
    
    return y_predict, caches






y_predict, caches = L_forward_propagate(x_train, parameters, layers)

print(y_predict)
assert(y_predict.shape == (1, 209))
```
